```python

print("Wavelet Transform analysis of Signals:::::::::::::::::::::::::::: \n");



import pywt
from matplotlib import pyplot as plt

print("Single Level Discrete Wavelet Transform and Inverse Transform :  \n\n");
x = [3,7,1,1, -2, 5,4 ,6]
print( "The Original signal: ", x);
 #signal
cA,cD = pywt.dwt(x,'haar') #wavelet transform

y = pywt.idwt(cA,cD, 'haar') #inverse wavelet transform


print("Printing Approximation coeffs : \n");
print(cA); #Printing Approximation coeffs
print("\n");
print("Printing Detailed coeffs : \n");
print(cD); #Printing Detailed Coeffs
print("\n");
print("Printing Reconstructed coeffs : \n");
print(y); #Printing Reconstructed Signal
print("\n");


plt.figure(figsize=(20,20));

plt.subplot(4,4,1)
plt.plot(cA,color = 'g');
plt.xlabel('Original signal');
plt.ylabel('cA');
plt.title("Approximation coeffs. (cA)");

print("\n");

plt.subplot(4,4,2)
plt.plot(y,color = 'b');
plt.xlabel('Time');
plt.ylabel('Value');
plt.title("Reconstructed Signal");

plt.savefig('single_level.png', dpi = 200)
plt.show()



print("\n");
print("\n");
print("##########################################################################\n");
print("\n");
print("Multilevel Level Discrete Wavelet Transform and Inverse Transform :  \n\n");
print("##########################################################################\n");
print("\n");


a=[3,7,1,1,-2,5,4,6,6,4,5,-2,1,1,7,3];
print("The Original signal:",a);


coeffs = pywt.wavedec(a,'db1', level=2, mode='periodic')
cA2,cD2,cD1 = coeffs;

b = pywt.waverec(coeffs,'db1', mode = 'periodic') #IDWT


print("\n");
print("Printing Approximation coeffs : \n");
print(cA2); #Printing Approximation coeffs
print("\n");

print("Printing Detailed coeffs D2 : \n");
print(cD2); #Printing Detailed Coeffs
print("\n");
print("Printing Detailed coeffs D1 : \n");
print(cD1); #Printing Detailed Coeffs
print("\n");
print("Printing Reconstructed coeffs : \n");
print(b); #Printing Reconstructed Signal
print("\n");


plt.figure(figsize=(20,20));

plt.subplot(4,4,1)
plt.plot(cA2,color = 'g');
plt.xlabel('Original signal');
plt.ylabel('cA2');
plt.title("Approximation coeffs. (cA2)");


plt.subplot(4,4,2)
plt.plot(cD2,color = 'b');
plt.xlabel('Original signal');
plt.ylabel('cD2');
plt.title("Detailed D2 coeffs. (cD2)");

plt.subplot(4,4,3)
plt.plot(cD1,color = 'r');
plt.xlabel('Original signal');
plt.ylabel('cD1');
plt.title("Detailed D1 coeffs. (cD1)");



plt.subplot(4,4,4)
plt.plot(b,color = 'y');
plt.xlabel('Time');
plt.ylabel('Value');
plt.title("Reconstructed Signal");

plt.savefig('multi_level.png', dpi = 200)
plt.show()





```
